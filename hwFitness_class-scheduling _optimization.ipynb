```python
# Fitness Studio Class Scheduling Optimization Project (Jupyter Notebook)

# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Defining the healthy weight prediction function from the model file
def healthy_weight_prediction(height, current_weight):
    height_m = height / 100  # converting height to meters
    min_bmi = 18.5
    max_bmi = 24.9
    min_healthy_weight = min_bmi * (height_m ** 2)
    max_healthy_weight = max_bmi * (height_m ** 2)
    return round(min_healthy_weight, 2), round(max_healthy_weight, 2)

# Sample data for class attendance (for optimization demo purposes)
class_data = {
    "class_time": ["6 AM", "8 AM", "10 AM", "12 PM", "2 PM", "4 PM", "6 PM", "8 PM"],
    "num_attendees": [10, 20, 15, 12, 30, 25, 40, 35]
}

# Create a DataFrame to analyze class attendance
attendance_df = pd.DataFrame(class_data)

# Visualize the class attendance
plt.figure(figsize=(10,6))
plt.bar(attendance_df["class_time"], attendance_df["num_attendees"], color="skyblue")
plt.xlabel("Class Time")
plt.ylabel("Number of Attendees")
plt.title("Class Attendance Analysis")
plt.show()

# Apply Linear Regression to predict future attendance based on the time of day
# Encode class time into numerical values for regression model (just a simple approach)
time_mapping = {
    "6 AM": 6, "8 AM": 8, "10 AM": 10, "12 PM": 12, "2 PM": 14, "4 PM": 16, "6 PM": 18, "8 PM": 20
}

attendance_df["time_num"] = attendance_df["class_time"].map(time_mapping)

# Train a simple linear regression model to predict attendees based on class time
X = attendance_df[["time_num"]]
y = attendance_df["num_attendees"]

model = LinearRegression()
model.fit(X, y)

# Predict future attendance at a given time
predicted_attendance = model.predict([[7]])  # Predict for a class at 7 AM
print(f"Predicted attendance at 7 AM: {predicted_attendance[0]} attendees")
```

### --- Used machine learning to analyse attendance data and customer preferences for optimal scheduling.

## Labeling Attendance Levels


```python
##Convert the Attendees data into three categories: Low, Medium, and High.

import pandas as pd

# Load your preprocessed data
df_cleaned = pd.read_excel("C:/Users/munny/preprocessed_fitness_studio_data.xlsx")

# Label the 'Attendees' column as 'Low', 'Medium', 'High' based on quantiles
attendance_labels = pd.qcut(df_cleaned['Attendees'], q=3, labels=['Low', 'Medium', 'High'])
df_cleaned['Attendance Level'] = attendance_labels
```

## Features and Target for Random Forest Classifier

```python
## Defining features and target variable.

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

# Define features and target
features = ['Time of Day', 'Day of Week', 'Class Type', 'Customer Preference', 'Avg Age', 'Male %', 'Female %']
target = 'Attendance Level'

# Convert categorical columns to numerical
df_encoded = df_cleaned.copy()
for col in ['Time of Day', 'Day of Week', 'Class Type', 'Customer Preference']:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Split the data into training and testing sets
X = df_encoded[features]
y = df_encoded[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Building Random Forest Classifier

```python
##Fitting the model and generate predictions

# Initialize and fit the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Predict and generate a classification report
y_pred = rf_classifier.predict(X_test)
classification_report_rf = classification_report(y_test, y_pred)

print(classification_report_rf)

```

# K-means Clustering

```python
##Cluster customers based on their preferences and behaviors.

from sklearn.cluster import KMeans

# For clustering, use 'Avg Age', 'Male %', 'Female %', 'Time of Day', 'Class Type'
clustering_features = ['Avg Age', 'Male %', 'Female %', 'Time of Day', 'Class Type']
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(df_encoded[clustering_features])

# Add cluster labels to the dataframe
df_encoded['Customer Cluster'] = clusters

# Save the clustered results to Excel
df_encoded.to_excel('fitness_studio_modeling_results.xlsx', index=False)
```

## Analyzing the results, tuning the model, and visualizing the cluster

```python
## STEP 01:  Analyze the Results
## 1.1. Classification Report Analysis

print(classification_report_rf)
```

```python
## 1.2. Confusion Matrix

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Low', 'Medium', 'High'], yticklabels=['Low', 'Medium', 'High'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```

```python
## STEP 02: Hyperparameter Tuning
## 2.1. Using GridSearchCV -  To find the best hyperparameters for the Random Forest model, which can improve its performance.

from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
}

# Create GridSearchCV object
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42), param_grid=param_grid, cv=3)
grid_search.fit(X_train, y_train)

# Display best parameters
print("Best parameters:", grid_search.best_params_)
```

```python
## Step 3: Visualizing Clusters with K-means
## 3.1. Visualizing Clusters

from sklearn.decomposition import PCA

# Reduce dimensions to 2D using PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df_encoded[clustering_features])

# Create a DataFrame for plotting
df_pca = pd.DataFrame(reduced_data, columns=['PC1', 'PC2'])
df_pca['Cluster'] = clusters

# Plotting the clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='Cluster', palette='viridis', s=100)
plt.title('Customer Clusters Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```

```python
## Step 4: Interpretation of Results

# Get feature importance
importances = rf_classifier.feature_importances_
feature_importance_df = pd.DataFrame({'Feature': features, 'Importance': importances}).sort_values(by='Importance', ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importance_df, x='Importance', y='Feature')
plt.title('Feature Importance')
plt.show()
```
